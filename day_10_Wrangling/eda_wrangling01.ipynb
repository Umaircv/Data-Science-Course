## How to deal with missing values in EDA analysis?
## cataogorical /object, numeric and boolean
### Handling missing values is a crucial step in Exploratory Data Analysis (EDA). Let's break it down by data type:
## Remember : Catagorical values means yes/no, male/female in this case we cant take mean we can take mode.

1. **Numeric Values:**
   - **Imputation:** Replace missing numeric values with the mean, median, or mode of the column. This helps to maintain the overall distribution.
   - **Flagging:** Introduce a new boolean column indicating whether the value was missing in the original column. This can provide additional information.

2. **Categorical/Object Values:**
   - **Mode Imputation:** Replace missing categorical values with the mode (most frequent category).
   - **Creating a New Category:** Introduce a new category (e.g., "Unknown" or "Other") to represent missing values.
   - **Use Predictive Models:** Train a model to predict missing categorical values based on other features.

3. **Boolean Values:**
   - **Imputation:** Since boolean values are typically 0 or 1, you might replace missing values with the most frequent value (0 or 1).
   - **Flagging:** Similar to numeric values, you can create a new boolean column indicating whether the original value was missing.

Always remember to document and justify your choice of imputation method, as it can impact the results of your analysis.

## 1st Method- Replacing Nan values with mean.


```python
# replace null values with mean in age column
df1['age'].isnull().sum()
```

```python
print (df1['age'].max())
print (df1['age'].min())
print (df1['age'].mean())
```

```python
df1['age'].fillna(df1['age'].mean(), inplace=True)
df1
```

```python

# df1 has no null values replaced with mean
df1.isnull().sum()
```

```python
df2=df1
```

```python
df2.isnull().sum()
```

## 2nd method- If there is many nan values so we will drop the coulmn.
## e.g; deck

```python
## now we remove nan from deck column.
## as there is many nan values so we drop the coulmn.
df2['deck'].unique()
```

```python
## dropping deck column
df2=df2.drop(columns=["deck"])
```

```python
## after dropping we are checking null values.
df2.isnull().sum()
```

## Method 3- Replacing values with mode

```python
## now we replace nan values with mode in embarked and embarked town column.
df3=df2
```

```python
df3['embarked'].fillna(df3['embarked'].mode()[0], inplace=True)
df3['embark_town'].fillna(df3['embark_town'].mode()[0], inplace=True)
df3
```

```python
# there is no null values in df3
df3.isnull().sum()
```
